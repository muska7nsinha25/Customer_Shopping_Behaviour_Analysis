```python
# with which should we replace our null values by mean or by median 

# we should always replace the null values with the meadian because mean gets effected 
# by the outliers and that would skew the distribution of our data.
# we should choose median because it is robust to the outliers.
```

```python
# Imputing missing values in Review Rating column with the median rating of the product category

df["Review Rating"]= df.groupby("Category")["Review Rating"].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# Renaming columns according to snake casing for better readability and documentation

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ","_")
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# new column age_group
labels = ["Young Adult", "Adult", "Middle_Aged", "Senior"]
df["age_group"]= pd.qcut(df["age"],q=4,labels= labels)
```

```python
df[["age","age_group"]].head(10)
```

```python
# new column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}
df["purchase_frequency_days"] = df["frequency_of_purchases"].map(frequency_mapping)
```
